creating labels for resized images

```python
labels = []
filenames = os.listdir('/content/gdrive/MyDrive/Research_2/resized_data3/')

for i in range(len(filenames)):
  file_name = filenames[i]
  label = file_name[0:3]

  if label =='man':
    labels.append(1)
  elif label == 'jav':
    labels.append(2)
  elif label == 'ora':
    labels.append(3)
  elif label == 'jac':
    labels.append(4)
  else:
    labels.append(0)
```

```python
import cv2
import glob
```

```python
image_diroctory = '/content/gdrive/MyDrive/Research_2/resized_data3/'
image_extension = ['png','jpg','jpeg']

files = []

[files.extend(glob.glob(image_diroctory +'*.' + e)) for e in image_extension]

dataset_images = np.asarray([cv2.imread(file) for file in files])
```

```python
print(dataset_images)
print(len(dataset_images))
```

```python
X = dataset_images
y = np.asarray(labels)
#y = y.reshape(-1)
```

```python
print(len(y))
print(len(dataset_images))
```

Train-Test split

```python
X_train,X_test,y_train, y_test = train_test_split(X, y, test_size = 0.4, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
X_train_scalled = X_train/255
X_test_scalled = X_test/255
```

```python
print(X_train_scalled)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Model

from tensorflow.keras import  layers, models
import matplotlib.pyplot as plt
from keras.layers.attention.multi_head_attention import activation

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

# DEEP LEARNING IMPORTS
from keras.models import Sequential
from keras.layers import Dense, Conv2D, Activation,Flatten, MaxPooling2D
```

```python

cnn = models.Sequential([
    layers.Conv2D(filters = 32, kernel_size = (3,3), activation = 'relu', input_shape = (32,32,3)),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(filters = 32, kernel_size = (3,3), activation = 'relu', input_shape = (32,32,3)),
    layers.MaxPooling2D((2,2)),

    layers.Flatten(),
    layers.Dense(64, activation = 'relu'),
    layers.Dense(10, activation = 'softmax')


])
```

```python
from sklearn import metrics
cnn.compile(optimizer='adam',
            loss = 'sparse_categorical_crossentropy',
            metrics = ['Accuracy']
            )
```

```python
cnn.fit(X_train_scalled,y_train, epochs=10)
```

```python
cnn.evaluate(X_test_scalled,y_test)
```

```python
from matplotlib.pyplot import *
import matplotlib.pyplot as plt
```

```python
y_test = y_test.reshape(-1)
```

```python
y_pred = cnn.predict(X_test_scalled)
y_pred[:5]
```

```python
y_classes = [np.argmax(element) for element in y_pred]
```

```python
print('Classification Report:\n', classification_report(y_test,y_classes))
```

```python
import tensorflow_hub as hub
```

```python
mobilenet_model = "https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4"

pretrained_model = hub.KerasLayer(mobilenet_model, input_shape = (224,224,3), trainable= False)
```

```python
num_of_classes = 3

model = tf.keras.Sequential([
    pretrained_model,
    tf.keras.layers.Dense(num_of_classes)
])

model.summary()
```

```python
model.compile(
    optimizer = 'adam',
    loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics = ['acc']
)
```

```python
model.fit(X_train_scalled, y_train, epochs=5)
```

```python
score,acc = model.evaluate(X_test_scalled,y_test)
print('Test Loss=', score)
print('Test Accuracy =', acc)
```
